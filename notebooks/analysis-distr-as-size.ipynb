```python
import pandas as pd
import numpy as np
from distfit import distfit
```

# Loading/preparing data set

```python
nodup_ips_pdf = pd.read_csv("output/202411_goscanner_ipv4_netstats_ip2location_asn_size.csv").drop_duplicates(subset=["ip"]).copy()[["nr_ips"]]
nodup_ips_pdf.sort_values(['nr_ips'], ascending=True, inplace=True)
```

```python
aggr_pdf = nodup_ips_pdf.groupby("nr_ips").size().reset_index(name="count")
```

# Test distribution (goodness-of-fit test)
